
# **Destino de proceso (compute target)**

Azure Machine Learning admite varios tipos de proceso para la experimentacin, el entrenamiento y la implementacin. Al tener varios tipos de proceso, puedes seleccionar el tipo de destino de proceso ms adecuado para tus necesidades concretas.



- **Instancia de proceso**: se comporta de forma similar a una mquina virtual y se usa principalmente para ejecutar cuadernos. Es idneo para la experimentacin.
- **Clsteres de proceso**: clsteres de varios nodos de mquinas virtuales que se escalan o reducen verticalmente de forma automtica para satisfacer la demanda. Una manera rentable de ejecutar scripts que necesitan procesar grandes volmenes de datos. Los clsteres tambin permiten usar el procesamiento paralelo para distribuir la carga de trabajo y reducir el tiempo necesario para ejecutar un script.
- **Clsteres de Kubernetes**: clster basado en la tecnologa Kubernetes, que proporciona ms control sobre cmo se configura y administra el proceso. Puede asociar el clster de Azure Kubernetes (AKS) autoadministrado para el proceso en la nube, o bien un clster Arc de Kubernetes para cargas de trabajo locales.
- **Proceso adjunto**: Permite asociar un proceso existente como mquinas virtuales de Azure o clsteres de Azure Databricks al rea de trabajo.
- **Proceso sin servidor**: Un proceso totalmente administrado a peticin que puede usar para los trabajos de entrenamiento.

## **Casos de Uso para Destinos de Proceso en Azure Machine Learning**

### **1. Eleccin de un Destino de Proceso para la Experimentacin**
- **Escenario:**  
  Durante la fase de experimentacin, los cientficos de datos suelen trabajar con pequeos subconjuntos de datos y requieren flexibilidad para probar diferentes enfoques rpidamente.
  
- **Opciones Recomendadas:**
  - **Instancia de Proceso (Compute Instance):**  
    Ideal para experimentacin en cuadernos Jupyter Notebook. Es una mquina virtual administrada por Azure que se ejecuta continuamente, permitiendo una experiencia fluida similar a trabajar localmente.
  - **Proceso sin Servidor de Spark:**  
    Si se necesita procesamiento distribuido o grandes volmenes de datos durante la experimentacin, puedes usar el proceso sin servidor de Spark para aprovechar la potencia de Apache Spark directamente en cuadernos.

- **Cundo Usarlo:**  
  - Cuando se necesita una configuracin rpida y fcil para experimentar con modelos.
  - Para trabajar en cuadernos interactivos con pequeos conjuntos de datos.

### **2. Eleccin de un Destino de Proceso para Produccin**
- **Escenario:**  
  Una vez finalizada la experimentacin, los scripts de entrenamiento deben automatizarse y escalarse para manejar grandes volmenes de datos en produccin.

- **Opciones Recomendadas:**
  - **Clster de Proceso (Compute Cluster):**  
    Un clster de proceso es ideal para ejecutar trabajos de entrenamiento a gran escala. Se escala automticamente segn la carga de trabajo y reduce los costos al apagarse cuando no est en uso.
  - **Proceso sin Servidor de Azure Machine Learning:**  
    Una alternativa administrada que elimina la necesidad de crear y gestionar clsteres manualmente, ideal para usuarios que prefieren una solucin completamente administrada.

- **Cundo Usarlo:**  
  - Cuando se necesita escalabilidad para manejar grandes volmenes de datos.
  - Para automatizar y programar trabajos de entrenamiento recurrentes.

### **3. Eleccin de un Destino de Proceso para el Despliegue**
- **Escenario:**  
  Despus de entrenar un modelo, se necesita implementarlo para generar predicciones. El tipo de despliegue depende de si las predicciones sern por lotes o en tiempo real.

#### **3.1. Predicciones por Lotes**
- **Opciones Recomendadas:**
  - **Clster de Proceso:**  
    Ideal para ejecutar trabajos de canalizacin que procesan grandes volmenes de datos en lotes.
  - **Proceso sin Servidor de Azure Machine Learning:**  
    Una alternativa escalable y administrada para ejecutar trabajos de prediccin por lotes sin necesidad de gestionar infraestructura.

- **Cundo Usarlo:**  
  - Cuando se necesita procesar grandes volmenes de datos en segundo plano sin interaccin en tiempo real.

#### **3.2. Predicciones en Tiempo Real**
- **Opciones Recomendadas:**
  - **Puntos de Conexin en Lnea Administrados:**  
    Azure ML crea y administra contenedores ligeros para implementar modelos como servicios RESTful, lo que permite predicciones en tiempo real con bajo costo.
  - **Azure Kubernetes Service (AKS):**  
    Ideal para implementaciones en tiempo real que requieren alta disponibilidad, escalabilidad y control total sobre el entorno de despliegue.

- **Cundo Usarlo:**  
  - Para aplicaciones que requieren respuestas en tiempo real, como recomendaciones personalizadas o deteccin de fraudes.
  - Cuando se necesita un despliegue escalable y robusto.

## **Creacin y uso de un clster de proceso**

Despus de la experimentacin y el desarrollo, se quiere que el cdigo est listo para produccin. 
- Al ejecutar cdigo en entornos de produccin, es mejor usar scripts en lugar de cuadernos. 
- Al ejecutar un script, querr usar un destino de proceso que sea escalable.

En Azure Machine Learning, los clsteres de proceso son idneos para ejecutar scripts. 
- Se puede crear un clster de proceso en Estudio de Azure Machine Learning, mediante la interfaz de la lnea de comandos (CLI) de Azure o el kit de desarrollo de software (SDK) de Python.


Al crear un clster de proceso, hay tres parmetros principales que debe tener en cuenta:

- **size**: especifica el tipo de mquina virtual de cada nodo dentro del clster de proceso. Se basa en los tamaos de las mquinas virtuales en Azure. Junto al tamao, tambin puede especificar si quiere usar CPU o GPU.
- **max_instances**: especifica el nmero mximo de nodos a los que el clster de proceso puede escalar horizontalmente. El nmero de cargas de trabajo paralelas que puede controlar el clster de proceso es anlogo al nmero de nodos a los que puede escalar el clster.
- **tier**: Especifica si las mquinas virtuales sern de prioridad baja o dedicadas. El establecimiento de una prioridad baja puede reducir los costos, ya que la disponibilidad no est garantizada.

```python
from azure.ai.ml.entities import AmlCompute

cluster_basic = AmlCompute(
    name="cpu-cluster",
    type="amlcompute",
    size="STANDARD_DS3_v2",
    location="westus",
    min_instances=0,
    max_instances=2,
    idle_time_before_scale_down=120,
    tier="low_priority",
)
ml_client.begin_create_or_update(cluster_basic).result()
```

### **Uso de un clster de proceso**
Hay tres escenarios principales en los que puede usar un clster de proceso:

- Ejecutar un job de pipeline que ha compilado en el Diseador.
- Ejecutar un job de aprendizaje automtico automatizado.
- Ejecutar un script como job.

Por ejemplo, para ejecutar un script como un job de comando, se puede establecer el destino de proceso en el clster:

```python
from azure.ai.ml import command

# configure job
job = command(
    code="./src",
    command="python diabetes-training.py",
    environment="AzureML-sklearn-0.24-ubuntu18.04-py37-cpu@latest",
    compute="cpu-cluster",
    display_name="train-with-cluster",
    experiment_name="diabetes-training"
    )

# submit job
returned_job = ml_client.create_or_update(job)
aml_url = returned_job.studio_url
print("Monitor your job at", aml_url)
```

Despus de enviar un trabajo que usa un clster de proceso, el clster de proceso se escalar horizontalmente a uno o varios nodos. 
- El cambio de tamao tardar unos minutos y el trabajo comenzar a ejecutarse una vez que se aprovisionen los nodos necesarios. 
- Cuando el estado de un trabajo es en preparacin, significa que el clster de proceso se est preparando. 
- Cuando el estado es en ejecucin, el clster de proceso est listo y el trabajo se est ejecutando.
