### **Recomendaciones**
- Para **pruebas rápidas o desarrollo**, usa VMs económicas como `STANDARD_DS1_V2` o `STANDARD_F2S_V2`.
- Para **cargas de trabajo generales**, considera opciones como `STANDARD_E2S_V3` o `STANDARD_D2AS_V4`.
- Para **aprendizaje profundo o inferencia de IA**, opta por VMs con GPUs como `STANDARD_NC4AS_T4_V3` o `STANDARD_NC24ADS_A100_V4`.

Los precios pueden variar según la región y la duración del uso. 
- Usar la [calculadora de precios de Azure](https://azure.microsoft.com/en-us/pricing/calculator/) para obtener estimaciones actualizadas.
